```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Openning dataset

```python
file = 'data_fCO2.txt'
data = pd.read_csv(file, sep='\t')
data
```

## Changing collum names

```python
data = data.rename(columns={'mon/day/yr': 'date', 'Longitude [degrees_east]':'longitude', 'Latitude [degrees_north]':'latitude'})
data
```

## Converting date to datetime format

```python
data['date'] = pd.to_datetime(data['date'].astype(str), format='%m/%d/%Y')
data
```

## First and Last data dates. Equal to get from other parameters

```python
print(data['date'].min())
print(data['date'].max())
```

## Selecting data by year, month or day using datetime

```python
data_2011 = data[pd.DatetimeIndex(data['date']).year == 2011]
#data_2011 = data[pd.DatetimeIndex(data['mon/day/yr']).month == 1]
#data_2011 = data[pd.DatetimeIndex(data['mon/day/yr']).day == 1]
data_2011
```

## Sorting dataset by somme collum

```python
data2 = data.sort_values(by=['date'], ascending=False, na_position='first')
data2
```

## Subsample dataset

```python
data3 = data[['longitude','latitude','SST','fCO2-T','fCO2-bio']]
data3
```

## Selecting and Counting unique values

```python
lons = data3['longitude'].unique()
lats = data3['latitude'].unique()

print(lons)
print(len(lons))
print(len(lats))
```

## Fltering and studing by unique lat and lons
## creating a new dataframe for the results

```python
dict_unique = {
    'lat': [],
    'lon': [],
    'avgSST': [],
    'diffMaxMin_fCO2-T': [],
    'diffMaxMin_fCO2-bio': []
}

for i in range (0, len(lats)):
    
    ## For each unique lat and lon
    data5 = data3[(data3['longitude']==lons[i]) & (data3['latitude']==lats[i])]
    v1 = round(np.max(data5['fCO2-T']) - np.min(data5['fCO2-T']),3)
    v2 = round(np.max(data5['fCO2-bio']) - np.min(data5['fCO2-bio']),3)
    dict_unique['lat'].append(lats[i])
    dict_unique['lon'].append(lons[i])
    dict_unique['avgSST'].append(round(np.mean(data5['SST']),3))
    dict_unique['diffMaxMin_fCO2-T'].append(v1)
    dict_unique['diffMaxMin_fCO2-bio'].append(v2)
```

```python
fig,ax = plt.subplots()
ax.scatter(range(0, len(lats)), dict_unique['diffMaxMin_fCO2-T'])
ax2=ax.twinx()
ax2.plot(dict_unique['avgSST'], color='red')
```

```python
fig,ax = plt.subplots()
ax.scatter(range(0, len(lats)), dict_unique['diffMaxMin_fCO2-bio'])
ax2=ax.twinx()
ax2.plot(dict_unique['avgSST'], color='red')
```
